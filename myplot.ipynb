```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
import pandas as pd
df=pd.read_excel('成绩表.xlsx')
df
x=df['姓名']
y1=df['数学']
y2=df['语文']
y3=df['英语']

plt.plot(x,y1,label='数学',color='r',marker='o')
plt.plot(x,y2,label='语文',color='g',marker='o',linestyle='--')
plt.plot(x,y3,label='英语',color='b',marker='o',linestyle='-.')

#设置网格线
plt.grid(axis='y')

#设置坐标轴的标题
plt.ylabel('分数')
plt.xlabel('姓名')

#设置图例
plt.legend(['数学','语文','英语'])

#设置坐标轴刻度
plt.yticks(range(50,150,10))
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']
import pandas as pd
df=pd.read_excel('产品销售统计.xlsx')
df

# #设置画布大小
plt.figure(figsize=(10,6))
x=df['产品名称']
height=df['总量']
plt.grid(axis='y',linestyle='--')

#设置x,y轴标签
plt.xlabel('产品名称',fontsize=15)
plt.ylabel('销量',fontsize=15)

#设置图标的标题
plt.title('2020年度产品销售量',fontsize=18)

#设置文本标签
for a,b in zip(x,height):
    plt.text(a,b,format(b,','),ha='center',va='bottom',fontsize=12,color='b',alpha=0.9)

plt.bar(x,height,width=0.5,alpha=0.5)
plt.legend(('销售额',),fontsize=15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.rcParams['font.sans-serif']=['SimHei']
df=pd.read_excel('产品销售统计.xlsx')

# #设置画布大小
plt.figure(figsize=(10,6))
x=np.array([0,1,2,3,4,5,6,7])
y1=df['1月']+df['2月']+df['3月']
y2=df['4月']+df['5月']+df['6月']
y3=df['7月']+df['8月']+df['9月']
y4=df['10月']+df['11月']+df['12月']

bar_width=0.2

#设置x,y轴的标签
plt.ylabel('季度销售量')
plt.xlabel('产品名称')

#设置图标的标题
plt.title('2021年季度销售量')

#绘制柱状图
plt.bar(x,y1,bar_width,color='c',alpha=0.5)
plt.bar(x+bar_width,y2,bar_width,color='b',alpha=0.5)
plt.bar(x+2*bar_width,y3,bar_width,color='g',alpha=0.5)
plt.bar(x+3*bar_width,y4,bar_width,color='r',alpha=0.5)

#设置坐标图刻度
data=df['产品名称']
plt.xticks(x,data)

#添加文本标签
for a,b in zip(x,y1):
    plt.text(a,b,format(b,','),ha='center',va='bottom',fontsize=8)
    
for a,b in zip(x,y2):
    plt.text(a+bar_width,b,format(b,','),ha='center',va='bottom',fontsize=8)
    
for a,b in zip(x,y3):
    plt.text(a+2*bar_width,b,format(b,','),ha='center',va='bottom',fontsize=8)
    
for a,b in zip(x,y4):
    plt.text(a+3*bar_width,b,format(b,','),ha='center',va='bottom',fontsize=8)
    
#设置图例
plt.legend(['第一季度','第二季度','第三季度','第四季度'])
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['font.sans-serif']=['SimHei']
df=pd.read_excel('JD手机销售量.xlsx')
df
#设置画布大小
plt.figure(figsize=(10,6))

x=df['北京出库量']
labels=df['商品名称']

plt.pie(x,labels=labels,autopct='%1.1f%%',labeldistance=1.02,startangle=45,textprops={'fontsize':12})

#设置x,y轴刻度一致，保证饼形图为圆形
plt.axis('equal')
plt.title('2021年4月京东手机品牌出库销售分析',fontsize=18)
plt.show()
```
